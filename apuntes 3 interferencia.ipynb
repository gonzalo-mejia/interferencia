```python
from pylab import plot, imshow, show, xlabel, ylabel
from numpy import loadtxt, array, zeros, sqrt, pi, sin, linspace
```

```python
cita=1
lamda=5
k=(2.*pi)/lamda
x1=0
y1=0
nx=500
dimx=100
dltx=dimx/nx

x=linspace(0,500,1)
y=linspace(0,500,1)
xlim(0.,500)
ylim(0.,500)

salida=open("ondas_est.dat","w")

for i in range(0,501) :
    for j in range(0,501):
        r1=sqrt((x1-i*dltx)**2+(y1-j*dltx)**2)
        C=cita*sin(k*r1)
        salida.write("%0.2f "% (cita*sin(k*r1)))
    salida.write("\n")
salida.close()
print(C)
data = loadtxt("ondas_est.dat", float)
imshow(data,origin="lower", extent=[0,100,0,100])
xlabel("Interferencia en X cm")
ylabel("interferencia en Y cm")
print(data.shape)
imshow(data)
show()

```
